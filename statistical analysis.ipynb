```r
library(gdata)
library(lme4)
mydata = read.csv("r_analysis.csv")

## lm normal
#lmodel <- lm('recuerdo~tiempo', mydata)
#summary(lmodel)
```

## memory decay

```r
## model con random effects (binomial)
m <- glmer(recuerdo~tiempo +
    (1 | subj), data = mydata, family = binomial, control = glmerControl(optimizer = "bobyqa"),
    nAGQ = 10)

# print the mod results without correlations among fixed effects
summary(m)
```

## memory decay by trick

```r
for(i in 1:8) {
    criteria <- c("juego == i")
    jx <-subset(mydata, eval(parse(text = criteria)))
    m <- glmer(recuerdo~session_o*tiempo +
        (1 | subj), data = jx, family = binomial, control = glmerControl(optimizer = "bobyqa"),
        nAGQ = 10)

    # print the mod results without correlations among fixed effects
    print(i)
    print(summary(m, corr = FALSE))
}
```
